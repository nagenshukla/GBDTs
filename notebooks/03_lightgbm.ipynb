# 03 · LightGBM - leaf-wise & histogram-based

LightGBM's speed comes from three ideas:

1. **Histogram binning.** Continuous features are bucketed into ~255 bins, so
   split-finding scans bins instead of sorted values - far less work.
2. **Leaf-wise growth.** Instead of growing all nodes at a level, LightGBM splits
   the single leaf that reduces loss most. Deeper, fewer-node trees - controlled
   by **`num_leaves`** rather than `max_depth`.
3. **GOSS & EFB.** Gradient-based One-Side Sampling keeps high-gradient rows;
   Exclusive Feature Bundling packs sparse features together.

```python
import os, sys, warnings, time
warnings.filterwarnings("ignore")
sys.path.insert(0, os.path.abspath(".."))
import numpy as np, pandas as pd, matplotlib.pyplot as plt
pd.set_option("display.precision", 3)
from utils import data as D
from utils import plotting as P
from sklearn.metrics import root_mean_squared_error, r2_score

df = pd.read_parquet("../data/regression.parquet")
y = df.pop("target")
X = df                                        # keeps 'category' as a categorical dtype
Xtr, Xte, ytr, yte = D.train_test_split_df(X, y, test_size=0.25, seed=0)
```

```python
from lightgbm import LGBMRegressor
import lightgbm as lgb

# LightGBM handles the categorical dtype natively.
model = LGBMRegressor(
    n_estimators=500, learning_rate=0.05, num_leaves=31,
    subsample=0.8, colsample_bytree=0.8, random_state=0, verbose=-1,
)
model.fit(Xtr, ytr, eval_set=[(Xtr, ytr), (Xte, yte)], eval_metric="rmse",
          callbacks=[lgb.early_stopping(30, verbose=False),
                     lgb.log_evaluation(0)])
print("best iteration:", model.best_iteration_)
print("test RMSE:", round(root_mean_squared_error(yte, model.predict(Xte)), 3),
      "| R2:", round(r2_score(yte, model.predict(Xte)), 3))
```

```python
hist = model.evals_result_
P.plot_training_curve({"train": hist["training"]["rmse"],
                       "validation": hist["valid_1"]["rmse"]},
                      title="LightGBM: RMSE vs iteration")
plt.show()
```

## `num_leaves`: the key knob

Leaf-wise growth is powerful but can overfit if `num_leaves` is too large
relative to the data. Sweep it.

```python
rows = []
for nl in [7, 15, 31, 63, 127, 255]:
    m = LGBMRegressor(n_estimators=300, learning_rate=0.05, num_leaves=nl,
                      random_state=0, verbose=-1)
    m.fit(Xtr, ytr)
    rows.append({"num_leaves": nl,
                 "train_RMSE": root_mean_squared_error(ytr, m.predict(Xtr)),
                 "val_RMSE": root_mean_squared_error(yte, m.predict(Xte))})
res = pd.DataFrame(rows).set_index("num_leaves")
print(res.round(3))
res.plot(marker="o", title="LightGBM: train vs val RMSE as num_leaves grows")
plt.ylabel("RMSE"); plt.grid(alpha=0.3); plt.show()
```

## Importance & speed

```python
_, _, gt = D.make_dataset("regression", n_samples=4000, seed=42)
imp = dict(zip(X.columns, model.feature_importances_))
P.plot_importance_vs_truth(imp, gt, model_name="LightGBM (split count)")
plt.show()
```

```python
# Quick speed sanity check vs a larger synthetic set.
Xbig, ybig, _ = D.make_dataset("regression", n_samples=100_000, seed=1)
t0 = time.perf_counter()
LGBMRegressor(n_estimators=300, num_leaves=31, verbose=-1).fit(Xbig, ybig)
print(f"LightGBM trained 300 trees on 100k rows in {time.perf_counter()-t0:.2f}s")
```

## Takeaways
- Histogram binning + leaf-wise growth = very fast, very accurate.
- Tune **`num_leaves`** (and `min_child_samples`) to control overfitting.
- Native categorical support; no manual encoding needed.
